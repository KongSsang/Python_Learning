#### 문제 4.
다음 점수 기준에 따라 Grade가 정해지는 코드를 완성하세요.

90점 이상: A
80점 이상: B
70점 이상: C
나머지: F

```
score = 60

if score >= 90:
    print('A')
elif score >= 80:
    print('B')
elif score >= 70:
    print('C')
else:
    print('F')
```

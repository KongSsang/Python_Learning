**선 그래프 그리기**

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

**산점도 그리기**

```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

**넘파이 배열로 산점도 그리기**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```

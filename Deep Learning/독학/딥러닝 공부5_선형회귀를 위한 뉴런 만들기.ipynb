```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
x = diabetes.data[:, 2]
y = diabetes.target
class Neuron:

  def __init__(self): #초기화 작업 수행
    self.w = 1.0
    self.b = 1.0

  def forpass(self, x): #정방향 계산 만들기
    y_hat = x * self.w + self.b
    return y_hat
  
  def backprop(self, x, err): #역방향 계산 만들기
    w_grad = x * err
    b_grad = 1 * err
    return w_grad, b_grad

  def fit(self, x, y, epochs = 100): # 훈련을 위한 메소드
    for i in range(epochs):
      for x_i, y_i in zip(x,y):
        y_hat = self.forpass(x_i)
        err = -(y_i - y_hat)
        w_grad, b_grad = self.backprop(x_i, err)
        self.w -= w_grad
        self.b -= b_grad

neuron = Neuron() # 모델 훈련하기
neuron.fit(x, y)

plt.scatter(x, y) # 학습 완료된 모델 가중치와 절편 확인
pt1 = (-0.1, -0.1 * neuron.w + neuron.b)
pt2 = (0.15, 0.15 * neuron.w + neuron.b)
plt.plot([pt1[0] + pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
